```python
df1["Course Sylabus"] = df1["Course Sylabus"].replace(["", "NA", "NaN", 0], np.nan)
```

```python
df1 = df1.dropna(subset=["Course Sylabus"])
df1
```

```python
df1["Concept"] = df1["Concept"].replace(["", "NA", 0], np.nan)
df1
```

```python
df1 = df1.dropna(subset=["Concept"])
df1
```

```python
df1["Course Sylabus"] = df1["Course Sylabus"].str.replace(r"<br\s*/?>", "", regex=True)
df1
```

```python
df1=df1[['Course Sylabus' , 'Concept']]
df1
```

```python
unique_course_syllabus = df1[["Course Sylabus"]].drop_duplicates()
unique_course_syllabus
```

```python
# Extract relevant columns
df3 = unique_course_syllabus["Course Sylabus"]
df4 = df1[["Concept", "Course Sylabus"]]
```

```python
# Merge df3 with df4 based on 'Unique_id'
merged_df = pd.merge(df3, df4, on="Course Sylabus", how="left")
```

```python
# Group by 'Unique_id' and aggregate 'Skill_conc' values into a list
grouped_df = merged_df.groupby("Course Sylabus")["Concept"].apply(list).reset_index()
grouped_df
```

```python
# Convert the list of skill_conc to a string for each Unique_id
# grouped_df['id'] = grouped_df['id'].apply(lambda x: ', '.join(map(str, x)))
grouped_df['Concept'] = grouped_df['Concept'].apply(lambda x: str(x))
grouped_df['Concept'].astype(str)
grouped_df
```

```python
grouped_df['Concept'] = grouped_df['Concept'].str.translate(str.maketrans({"[": "", "]": ""}))
```

```python
grouped_df['Concept'] = grouped_df['Concept'].str.replace(".0", "", regex=False)
grouped_df
```

```python
def remove_duplicates_from_string(cell_value):
    # Split the string by commas and strip any spaces
    values = cell_value.split(", ")
    
    # Remove duplicates by converting to a set and back to a list
    unique_values = list(set(values))
    
    # Join the unique values back into a string, separating by commas
    return ", ".join(unique_values)

# Example usage:
grouped_df['Concept'].apply(remove_duplicates_from_string)
grouped_df
```

```python
grouped_df["Course Sylabus"] = (
    grouped_df["Course Sylabus"]
    .astype(str)
    .str.encode("ascii", "ignore")
    .str.decode("ascii")
)

```

```python
def remove_duplicates_in_cell(x):
    # 1. Handle missing values
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return x
    
    # 2. If value is already a list or array
    if isinstance(x, (list, tuple)):
        items = [str(i).strip() for i in x]
    else:
        # 3. Otherwise treat it as a string
        items = [i.strip() for i in str(x).split(',')]
    
    # 4. Remove duplicates while keeping order
    unique_items = list(dict.fromkeys(items))
    
    # 5. Join back into a clean string
    return ", ".join(unique_items)

# Apply safely
grouped_df['Concept'] = grouped_df['Concept'].apply(remove_duplicates_in_cell)

```
