#### ###CONCEPTS_ID SEPARATE BY ROW


```python
import pandas as pd

# Load the Excel file
file_path = r"C:\Users\taslim.siddiqui\OneDrive - Eduvanz Financing Private Limited\Desktop\RECO RATING\new  04-11-2024\Recorating_13-10-2025.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Drop rows where 'concepts_id' is NaN or empty
df = df.dropna(subset=['concepts'])

# Convert 'concepts_id' column to string and split by comma
df['concepts'] = df['concepts'].astype(str).str.split(r'\s*,\s*')

# Explode the concepts_id column
df_exploded = df.explode('concepts').reset_index(drop=True)

# Remove rows where exploded 'concepts_id' is empty string or NaN
df_exploded = df_exploded[df_exploded['concepts'].str.strip().astype(bool)]

# Rename 'concepts_id' to 'concept'
df_exploded.rename(columns={'concepts': 'concepts_id'}, inplace=True)

# Move 'concept' column to the first position
cols = ['concepts_id'] + [col for col in df_exploded.columns if col != 'concepts_id']
df_exploded = df_exploded[cols]

# Save to a new Excel file
output_path = r"C:\Users\taslim.siddiqui\OneDrive - Eduvanz Financing Private Limited\Desktop\RECO RATING\old_13-11-2025_output.xlsx"
df_exploded.to_excel(output_path, index=False)

print(f"File saved successfully as {output_path}")
```

## #Merge the two file on concept_id


```python
import pandas as pd

# Load both Excel files
df_expanded_concepts = pd.read_excel(
    r"C:\Users\taslim.siddiqui\OneDrive - Eduvanz Financing Private Limited\Desktop\RECO RATING\new  04-11-2024\new  03-11-2025_output.xlsx"
)
df_abhfl_abm_mapping = pd.read_excel(
    r"C:\Users\taslim.siddiqui\Downloads\Master_ABC.xlsx"
)

# Drop missing Concept IDs
df_expanded_concepts = df_expanded_concepts.dropna(subset=["concepts_id"])
df_abhfl_abm_mapping = df_abhfl_abm_mapping.dropna(subset=["Concepts_id"])

# Convert Concepts_id columns to numeric safely
df_expanded_concepts["concepts_id"] = pd.to_numeric(df_expanded_concepts["concepts_id"], errors="coerce")
df_abhfl_abm_mapping["Concepts_id"] = pd.to_numeric(df_abhfl_abm_mapping["Concepts_id"], errors="coerce")

# Drop rows that couldn't be converted
df_expanded_concepts = df_expanded_concepts.dropna(subset=["concepts_id"])
df_abhfl_abm_mapping = df_abhfl_abm_mapping.dropna(subset=["Concepts_id"])

# Convert back to int
df_expanded_concepts["concepts_id"] = df_expanded_concepts["concepts_id"].astype(int)
df_abhfl_abm_mapping["Concepts_id"] = df_abhfl_abm_mapping["Concepts_id"].astype(int)

# Filter only matching Concept IDs before merge
df_expanded_concepts_filtered = df_expanded_concepts[
    df_expanded_concepts["concepts_id"].isin(df_abhfl_abm_mapping["Concepts_id"])
]

# Perform LEFT JOIN
merged_output = df_expanded_concepts_filtered.merge(
    df_abhfl_abm_mapping,
    left_on="concepts_id",
    right_on="Concepts_id",
    how="left"
)

# Save final output
output_file = r"C:\Users\taslim.siddiqui\OneDrive - Eduvanz Financing Private Limited\Desktop\RECO RATING\new  04-11-2024\new_04-11-2025_output_explaned.xlsx"
merged_output.to_excel(output_file, index=False)

print("✅ Final matched file saved at:", output_file)
```
