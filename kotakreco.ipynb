```python
import pandas as pd
```

```python
df = pd.read_excel("/Users/romitbenkar/Downloads/Additional Skills and Concepts upload - Kotak HR (Locked FSC).xlsx",sheet_name="Sheet1")
df
```

```python
df.rename(columns={'Duration': 'duration_medium__c'},inplace=True)
df.columns
```

```python
# Replace 'apple' with 'orange' and 'banana' with 'grape' in column 'B'
df['duration_medium__c'] = df['duration_medium__c'].replace({'Day': 'Days', 'Month': 'Months','Year':'Years'})

# Define conversion factors to seconds for each mode
conversion_factors = {
    'years': 365 * 24 * 60 * 60,        # approximate (ignores leap years)
    'months': 30.44 * 24 * 60 * 60,     # average month
    'weeks': 7 * 24 * 60 * 60,
    'days': 24 * 60 * 60,
    'hours': 60 * 60,
    'minutes': 60
}

df['duration_medium__c']=df['duration_medium__c'].astype(str)


# Function to convert duration based on the mode
def convert_to_seconds(row):
    duration = row['course_duration__c']
    mode = row['duration_medium__c'].strip().lower()  # normalize to lowercase
    return duration * conversion_factors.get(mode, 1)  # Default to 1 if mode is not found

# Apply the conversion function to each row
df['Duration_in_seconds'] = df.apply(convert_to_seconds, axis=1)

# Add a column that checks if the duration in seconds is less than or equal to a week
df['Is_within_a_week'] = df['Duration_in_seconds'] <= 7 * 24 * 60 * 60

# Save the result back to Excel
#df.to_excel('//Users//romitbenkar//Desktop//Office//19-11-24//timeconvertoutput.xlsx', index=False)
df
```

```python
df.rename(columns={'Institute': 'CPP'},inplace=True)
df.columns
```

```python
# Define a function to count the number of concepts based on commas
def count_commas(val):
    if pd.isna(val):  # If the value is NaN (missing), return 0
        return 0
    elif ',' in val:  # If the value contains commas, count the number of commas and add 1
        return val.count(',') + 1
    else:  # If the value has no commas, assume it's a single concept
        return 1

# Apply the function to each value in the 'concepts' column
# and create a new column 'concept_count' with the result
df['concept_count'] = df['Concept id'].apply(count_commas)
```

```python
values_to_check_keka = ['Keka Academy']

# Check if any values in the DataFrame are in the list
df['integration_keka'] = df['CPP'].isin(values_to_check_keka)
integration_keka_df = df.query("integration_keka == True")
integration_keka_df
```

```python
values_to_check_HH = ['Henry Harvin']

# Check if any values in the DataFrame are in the list
df['integration_HH'] = df['CPP'].isin(values_to_check_HH)
integration_HH_df = df.query("integration_HH == True")
integration_HH_df
```

```python
values_to_check_Eduonix = ['Eduonix']

# Check if any values in the DataFrame are in the list
df['integration_Eduonix'] = df['CPP'].isin(values_to_check_Eduonix)
integration_Eduonix_df = df.query("integration_Eduonix == True")
integration_Eduonix_df
```

```python
rank_integration_keka_df = integration_keka_df.sort_values(
    by=["concept_count","Duration_in_seconds"],
    ascending=[False,True]
).reset_index(drop=True)


rank_integration_keka_df['Rank'] = 150 - (rank_integration_keka_df.index * 0.01)
rank_integration_keka_df
```

```python
rank_integration_HH_df = integration_HH_df.sort_values(
    by=["concept_count","Duration_in_seconds"],
    ascending=[False,True]
).reset_index(drop=True)


rank_integration_HH_df['Rank'] = 100 - (rank_integration_HH_df.index * 0.01)
rank_integration_HH_df
```

```python
rank_integration_Eduonix_df = integration_Eduonix_df.sort_values(
    by=["concept_count","Duration_in_seconds"],
    ascending=[False,True]
).reset_index(drop=True)


rank_integration_Eduonix_df['Rank'] = 50 - (rank_integration_Eduonix_df.index * 0.01)
rank_integration_Eduonix_df
```

```python
# Define a dictionary with keys representing category labels 
# and values being different filtered and ranked DataFrames
pieces = {
    "x1": rank_integration_keka_df,   # Non-corporate courses marked as priority_cpp
    "x2": rank_integration_HH_df,                # Non-corporate courses that have concepts
    "x3": rank_integration_Eduonix_df,              # Non-corporate courses with no concept mapping
}

# Concatenate all the DataFrames together vertically (i.e., one below the other)
result = pd.concat(pieces)

# Save the final concatenated DataFrame to an Excel file
# result.to_excel("/Users/romitbenkar/Downloads/Recorating_Kotak_HR.xlsx")
result
```
