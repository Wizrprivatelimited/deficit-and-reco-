```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np

# Define the cleaning function
def clean_syllabus(text):
    if pd.isnull(text) or str(text).strip().lower() == 'null' or str(text).strip() == '':
        return ''  # Keep as empty string instead of dropping
    # Replace <br> with line break and remove other HTML tags
    text_with_breaks = str(text).replace('<br>', '\n')
    soup = BeautifulSoup(text_with_breaks, "html.parser")
    return soup.get_text(separator="\n")

# Load the Excel file
file_path = "C:\\Users\\taslim.siddiqui\\Downloads\\product_data.xlsx"  # Adjust path as needed
df = pd.read_excel(file_path)

# Store original column for comparison
df['course_syllabus_old'] = df['course_syllabus__c']

# Apply the cleaning function to create a new column
df['course_syllabus_new'] = df['course_syllabus__c'].apply(clean_syllabus)

# Keep all rows, including blank/empty ones
df_result = df[['id', 'course_syllabus_old', 'course_syllabus_new']]

# Save the result to a new Excel file
output_path = "C:\\Users\\taslim.siddiqui\\Downloads\\cleaned_course_syllabus_comparison.xlsx"
df_result.to_excel(output_path, index=False)

print("✅ Cleaned file saved to:", output_path)
```

```python
import pandas as pd

# =====================================
# 1. Load Input Files
# =====================================
mpid_df = pd.read_excel(r"C:\Users\taslim.siddiqui\Downloads\textjoin_kotak_09-12-2025.xlsx")   # MPID list
master_df = pd.read_excel(r"C:\Users\taslim.siddiqui\Downloads\Reco file Assurance - Audit 1 (1).xlsx",
                          sheet_name="Sheet2")  # Course master file

# =====================================
# 2. Clean Column Names (Avoid issues)
# =====================================
master_df.columns = master_df.columns.str.strip().str.replace(" ", "_")

# =====================================
# 3. Filter only required MPIDs
# =====================================
filtered_df = master_df[master_df["Mpid"].isin(mpid_df["Mpid"])]

# =====================================
# 4. Group & Aggregate
# =====================================
final_df = filtered_df.groupby("Mpid").agg({
    "Duration": "first",
    "Price": "first",
    "Course_Name": "first",
    "CPP": "first",                     # CPP exists in file
    "Course_Link": "first",
    "Course_id": lambda x: ", ".join(x.astype(str).unique()),
    "Duration_c": "first"
}).reset_index()

# Rename columns exactly as required
final_df.rename(columns={
    "Course_id": "Concept id",
    "Duration_c": "Duration_Type"
}, inplace=True)

# =====================================
# 5. Save Output
# =====================================
output_path = r"C:\Users\taslim.siddiqui\Downloads\final_output.xlsx"
final_df.to_excel(output_path, index=False)

print("✅ Final merged file saved to:", output_path)
```
