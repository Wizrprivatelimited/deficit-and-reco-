```python
# Extract relevant columns
df3 = df1["Concat"]
df4 = df2[["Concat", "Proficiency"]]
```

```python
# Merge df3 with df4 based on 'Unique_id'
merged_df = pd.merge(df3, df4, on="Concat", how="left")
```

```python
# Group by 'Unique_id' and aggregate 'Skill_conc' values into a list
grouped_df = merged_df.groupby("Concat")["Proficiency"].apply(list).reset_index()
```

```python
# Convert the list of skill_conc to a string for each Unique_id
# grouped_df['id'] = grouped_df['id'].apply(lambda x: ', '.join(map(str, x)))
grouped_df['Proficiency'] = grouped_df['Proficiency'].apply(lambda x: str(x))
grouped_df['Proficiency'].astype(str)
grouped_df
```

```python
grouped_df['Proficiency'] = grouped_df['Proficiency'].str.translate(str.maketrans({"[": "", "]": ""}))
```

```python
grouped_df['Proficiency'] = grouped_df['Proficiency'].str.replace(".0", "", regex=False)
grouped_df.head()
```

```python
def remove_duplicates_from_string(cell_value):
    # Split the string by commas and strip any spaces
    values = cell_value.split(", ")
    
    # Remove duplicates by converting to a set and back to a list
    unique_values = list(set(values))
    
    # Join the unique values back into a string, separating by commas
    return ", ".join(unique_values)

# Example usage:
grouped_df['Proficiency'].apply(remove_duplicates_from_string)
grouped_df.head()
```

```python
# Assuming 'df_exploded' is the DataFrame you want to save
# This function splits the DataFrame into multiple sheets and saves to a single Excel file

def save_large_dataframe_to_excel(df, output_file, chunk_size=1000000):
    """
    Save a large dataframe to an Excel file with multiple sheets.
    
    :param df: The DataFrame to save
    :param output_file: Path to the output Excel file
    :param chunk_size: The number of rows per sheet
    """
    with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
        # Split the dataframe into chunks and write each chunk to a separate sheet
        for i in range(0, len(df), chunk_size):
            sheet_name = f"Sheet_{i // chunk_size + 1}"  # Create sheet names like Sheet_1, Sheet_2, etc.
            df_chunk = df.iloc[i:i + chunk_size]
            df_chunk.to_excel(writer, sheet_name=sheet_name, index=False)

    print(f"Data saved to {output_file} with {len(df) // chunk_size + 1} sheets.")

# Example usage:
# Assuming df_exploded is the DataFrame that you want to save
output_file = "/Users/romitbenkar/Downloads/IT File/large_dataframe.xlsx"
save_large_dataframe_to_excel(df_exploded, output_file)
```
