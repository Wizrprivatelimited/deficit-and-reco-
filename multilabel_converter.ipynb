```python
import pandas as pd

# Load file
df = pd.read_excel(r"C:\Users\taslim.siddiqui\Downloads\TEXTJOIN_output_concept.xlsx")

# Skill columns (all except Concept)
skill_cols = [col for col in df.columns if col != "Concept"]

# Convert Concept cell into list (for processing only)
df["Concept_list"] = (
    df["Concept"]
    .astype(str)
    .str.replace(r"[\[\]']", "", regex=True)
    .str.split(",")
)

# Strip spaces inside the list
df["Concept_list"] = df["Concept_list"].apply(lambda lst: [x.strip() for x in lst])

# Create 0/1 indicators
for col in skill_cols:
    df[col] = df["Concept_list"].apply(lambda skills: 1 if col in skills else 0)

# Remove helper list column (keep original Concept column)
df.drop(columns=["Concept_list"], inplace=True)

# Save
df.to_excel(
    r"C:\Users\taslim.siddiqui\Downloads\TEXTJOIN_output_Concept_ML1.xlsx",
    index=False
)

print("Done! Output created with original Concept + 0/1 columns.")
```
