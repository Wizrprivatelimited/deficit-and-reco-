```python
import pandas as pd

path = r"C:\Users\taslim.siddiqui\Downloads\KOTAK Course + Deficit Tracker_Data added 20.01 2.xlsx"

df = pd.read_excel(path, sheet_name='HH+Other CPP Master')

# --- PIVOT: Function x Proficiency, distinct count of concepts ---
pivot = pd.pivot_table(
    df,
    index=['Function'],
    columns=['course_level'],
    values='concepts',          # or 'Concept' if that is the column name
    aggfunc=pd.Series.nunique,  # distinct count
    fill_value=0
)

pivot_out = pivot.reset_index()

print(pivot_out.head())

pivot_out.to_excel(
    r"C:\Users\taslim.siddiqui\Downloads\pivot_1.xlsx",
    index=False
)
```
